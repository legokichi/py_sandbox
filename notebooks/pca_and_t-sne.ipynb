# PCA と t-SNE

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

digit image data - http://scikit-learn.org/stable/auto_examples/datasets/plot_digits_last_image.html

```python
digits = datasets.load_digits()
print(digits.data.shape)
print(digits.target.shape)
plt.figure(1, figsize=(3, 3))
plt.imshow(digits.images[-1], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

## PCAで可視化
* 速い

```python
%%time
X_reduced = PCA(n_components=2).fit_transform(digits.data)
print(X_reduced.shape)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target)
plt.colorbar()
```

## t-SNE で可視化
* 説明 - https://deepage.net/machine_learning/2017/03/08/tsne.html
* 時間とメモリを消費するのが難点
* scikit-learn - http://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
* perplexity: 考慮する最近傍の数(大きいほど時間かかる

```python
%%time
X_reduced = TSNE(perplexity=2, random_state=0).fit_transform(digits.data)
print(X_reduced.shape)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target)
plt.colorbar()
```

```python
%%time
X_reduced = TSNE(perplexity=8, random_state=0).fit_transform(digits.data)
print(X_reduced.shape)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target)
plt.colorbar()
```
